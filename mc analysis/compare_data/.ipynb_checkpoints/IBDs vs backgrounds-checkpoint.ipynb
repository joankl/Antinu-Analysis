```python
import colorsys
# Función para generar variantes más claras y más oscuras de un color
def generate_color_variants(base_color, num_variants, factor = 0.3):
    # Convertir el color base a valores de HSV
    r, g, b = base_color
    h, l, s = colorsys.rgb_to_hls(r, g, b)
    
    # Generar variantes
    variants = []
    for i in range(1, num_variants + 1):
        # Hacer el color más claro
        lighter = colorsys.hls_to_rgb(h, min(1, l * (1 + factor * i)), s)
        # Hacer el color más oscuro
        darker = colorsys.hls_to_rgb(h, max(0, l * (1 - factor * i)), s)
        variants.append(lighter)
        variants.append(darker)
        
    return variants

# Color base (rojo)
base_red = (1, 0, 0) # RGB para rojo puro
# Color base (azuk)
base_blue = (0, 0, 1) # RGB para azul puro
num_variants = 5

red_variants = generate_color_variants(base_red, num_variants)
blue_variants = generate_color_variants(base_blue, num_variants)
```

# 1 - Import the saved data

```python
main_rute = 'C:/Users/joanc/jupyter notebooks/Antineutrino Analysis'

#full data
reactor_nu_data = pd.read_csv(main_rute + '/Reactor Nu/data/full_data.csv').drop(columns=['Unnamed: 0'])
alpha_n_data = pd.read_csv(main_rute + '/alpha_n/data/full_data.csv').drop(columns=['Unnamed: 0'])
bipo214_data = pd.read_csv(main_rute + '/BiPo214/data/full_data.csv').drop(columns=['Unnamed: 0'])

#Dt and Dr info of selected events
reactor_nu_dt_dr = pd.read_csv(main_rute + '/Reactor Nu/data/Dr_Dt_0.8MeV_5500mm.csv').drop(columns=['Unnamed: 0'])
alpha_n_dt_dr = pd.read_csv(main_rute + '/alpha_n/data/Dr_Dt_0.8MeV_5500mm.csv').drop(columns=['Unnamed: 0'])
bipo214_dt_dr = pd.read_csv(main_rute + '/BiPo214/data/Dr_Dt_0.8MeV_5500mm.csv').drop(columns=['Unnamed: 0'])
```

## 1.1 - Extract the observables

```python
reactor_nu_dt_dr
```

```python
#Energy cut
en_cut_inf = 0
en_cut_sup = 8

#Radial cut
r_cut_sup = 5500

#Separate the prompt from the delay date by using the evIndex
reactor_prompt_data = reactor_nu_data[(reactor_nu_data['evIndex'] == 0) & (reactor_nu_data['energy'] >= en_cut_inf) & (reactor_nu_data['energy'] <= en_cut_sup) & (reactor_nu_data['posr'] <= r_cut_sup)].reset_index()
reactor_delay_data = reactor_nu_data[(reactor_nu_data['evIndex'] == 2) & (reactor_nu_data['energy'] >= en_cut_inf) & (reactor_nu_data['energy'] <= en_cut_sup) & (reactor_nu_data['posr'] <= r_cut_sup)].reset_index()

alpha_n_prompt_data = alpha_n_data[(alpha_n_data['evIndex'] == 0) & (alpha_n_data['energy'] >= en_cut_inf) & (alpha_n_data['energy'] <= en_cut_sup) & (alpha_n_data['posr'] <= r_cut_sup)].reset_index()
alpha_n_delay_data = alpha_n_data[(alpha_n_data['evIndex'] == 2) & (alpha_n_data['energy'] >= en_cut_inf) & (alpha_n_data['energy'] <= en_cut_sup) & (alpha_n_data['posr'] <= r_cut_sup)].reset_index()

bipo214_prompt_data = bipo214_data[(bipo214_data['evIndex'] == 0) & (bipo214_data['energy'] >= en_cut_inf) & (bipo214_data['energy'] <= en_cut_sup) & (bipo214_data['posr'] <= r_cut_sup)].reset_index()
bipo214_delay_data = bipo214_data[(bipo214_data['evIndex'] == 2) & (bipo214_data['energy'] >= en_cut_inf) & (bipo214_data['energy'] <= en_cut_sup) & (bipo214_data['posr'] <= r_cut_sup)].reset_index()

#Extract the energy of the prompt and delayed
reactor_promp_en = np.array(reactor_prompt_data['energy'])
reactor_delay_en = np.array(reactor_delay_data['energy'])

alpha_n_prompt_en = np.array(alpha_n_prompt_data['energy'])
alpha_n_delay_en = np.array(alpha_n_delay_data['energy'])

bipo214_prompt_en = np.array(bipo214_prompt_data['energy'])
bipo214_delay_en = np.array(bipo214_delay_data['energy'])

#Extract the pre-computed Dt and Dr
reactor_Dt = np.array(reactor_nu_dt_dr['Dt (mcs)'])
reactor_Dr = np.array(reactor_nu_dt_dr['Dr (mm)'])

alpha_n_Dt = np.array(alpha_n_dt_dr['Dt (mcs)'])
alpha_n_Dr = np.array(alpha_n_dt_dr['Dr (mm)'])

bipo214_Dt = np.array(bipo214_dt_dr['Dt (mcs)'])
bipo214_Dr = np.array(bipo214_dt_dr['Dr (mm)'])
```

# 4 - Use the Energy vs. Dt and Energy vs. Dr

Lets construct bi-dimensional histograms to compare reactor neutrinos with backgrounds

## 4.1 - Extract the binned Data

```python
#Select Dt and Dr limits
Dt_lim = 800
Dr_lim = 700

reactor_Dt_cut = reactor_Dt[reactor_Dt <= Dt_lim]
reactor_Dr_cut = reactor_Dr[reactor_Dr <= Dr_lim]

alpha_n_Dt_cut = alpha_n_Dt[alpha_n_Dt <= Dt_lim]
alpha_n_Dr_cut = alpha_n_Dr[alpha_n_Dr <= Dr_lim]

bipo214_Dt_cut = bipo214_Dt[bipo214_Dt <= Dt_lim]
bipo214_Dr_cut = bipo214_Dr[bipo214_Dr <= Dr_lim]

#data limiter: Since x and y must have the same lenght, lest apply a cut on elements, even if this looks ugly, fix latter.
lim_reactor_dt = reactor_Dt_cut.shape[0]
lim_reactor_dr = reactor_Dr_cut.shape[0]

lim_alpha_n_dt = alpha_n_Dt_cut.shape[0]
lim_alpha_n_dr = alpha_n_Dr_cut.shape[0]

lim_bipo214_dt = bipo214_Dt_cut.shape[0]
lim_bipo214_dr = bipo214_Dr_cut.shape[0]

#Redefine the arrays to contain a zero count at en_cut_inf

# Contruct the bi_dimensional histograms
bins = 40

#reactor data
reactor_2d_en_dt, xedges, yedges = np.histogram2d(x = reactor_promp_en[:lim_reactor_dt], y = reactor_Dt_cut, bins = [bins, bins])
reactor_2d_en_dr, xedges, yedges = np.histogram2d(x = reactor_promp_en[:lim_reactor_dr], y = reactor_Dr_cut, bins = [bins, bins])
#Normalization
norm_reactor_dt = mcolors.Normalize(vmin=np.min(reactor_2d_en_dt), vmax=np.max(reactor_2d_en_dt))
norm_reactor_dr = mcolors.Normalize(vmin=np.min(reactor_2d_en_dr), vmax=np.max(reactor_2d_en_dr))

#alpha-n data
alpha_n_2d_en_dt, xedges, yedges = np.histogram2d(x = alpha_n_prompt_en[:lim_alpha_n_dt], y = alpha_n_Dt_cut, bins = [bins, bins])
alpha_n_2d_en_dr, xedges, yedges = np.histogram2d(x = alpha_n_prompt_en[:lim_alpha_n_dr], y = alpha_n_Dr_cut, bins = [bins, bins])
#Normalization
norm_alpha_n_dt = mcolors.Normalize(vmin=np.min(alpha_n_2d_en_dt), vmax=np.max(alpha_n_2d_en_dt))
norm_alpha_n_dr = mcolors.Normalize(vmin=np.min(alpha_n_2d_en_dr), vmax=np.max(alpha_n_2d_en_dr))

#BiPo data
bipo214_2d_en_dt, xedges, yedges = np.histogram2d(x = bipo214_prompt_en[:lim_bipo214_dt], y = bipo214_Dt_cut, bins = [bins, bins])
bipo214_2d_en_dr, xedges, yedges = np.histogram2d(x = bipo214_prompt_en[:lim_bipo214_dr], y = bipo214_Dr_cut, bins = [bins, bins])
#Normalization
norm_bipo214_dt = mcolors.Normalize(vmin=np.min(bipo214_2d_en_dt), vmax=np.max(bipo214_2d_en_dt))
norm_bipo214_dr = mcolors.Normalize(vmin=np.min(bipo214_2d_en_dr), vmax=np.max(bipo214_2d_en_dr))
```

```python
save = True

fig, axes = plt.subplots(2, 3,  figsize=(25, 14))

#Plt limit axis
#energy:
max_reactor_en = max(reactor_promp_en[:lim_reactor_dt])
max_alpha_n_en = max(alpha_n_prompt_en[:lim_alpha_n_dt])
max_bipo_214_en = max(bipo214_prompt_en[:lim_bipo214_dt])

min_reactor_en = min(reactor_promp_en[:lim_reactor_dt])
min_alpha_n_en = min(alpha_n_prompt_en[:lim_alpha_n_dt])
min_bipo_214_en = min(bipo214_prompt_en[:lim_bipo214_dt])
#en_max = en_cut_inf
#en_min = en_cut_sup

titles = [r'Reactor $\bar{\nu}$', r'$\alpha$-n', r'BiPo-214']

cmap = cm.PuBuGn

axes[0,0].imshow(reactor_2d_en_dt, cmap = cmap, aspect = 'auto', extent = [0, Dt_lim, max_reactor_en, min_reactor_en], norm = norm_reactor_dt)
axes[0,1].imshow(alpha_n_2d_en_dt, cmap = cmap, aspect = 'auto', extent = [0, Dt_lim, max_alpha_n_en, min_alpha_n_en], norm = norm_alpha_n_dt)
axes[0,2].imshow(bipo214_2d_en_dt, cmap = cmap, aspect = 'auto', extent = [0, Dt_lim, max_bipo_214_en, min_bipo_214_en], norm = norm_bipo214_dt)

axes[1,0].imshow(reactor_2d_en_dr, cmap = cmap, aspect = 'auto', extent = [0, Dr_lim, max_reactor_en, min_reactor_en], norm = norm_reactor_dr)
axes[1,1].imshow(alpha_n_2d_en_dr, cmap = cmap, aspect = 'auto', extent = [0, Dr_lim, max_alpha_n_en, min_alpha_n_en], norm = norm_alpha_n_dr)
axes[1,2].imshow(bipo214_2d_en_dr, cmap = cmap, aspect = 'auto', extent = [0, Dr_lim, max_bipo_214_en, min_bipo_214_en], norm = norm_bipo214_dr)

for i, title_i in enumerate(titles):
    axes[0,i].set_title(title_i)
    axes[0,i].set_xlabel(r'$\Delta$t ($\mu$s)')
    axes[0,i].set_ylabel(r'Prompt Energy (MeV)')
    axes[0,i].set_ylim(en_cut_inf, en_cut_sup)
    
    axes[1,i].set_title(title_i)
    axes[1,i].set_xlabel(r'$\Delta$r (mm)')
    axes[1,i].set_ylabel(r'Prompt Energy (MeV)')
    axes[1,i].set_ylim(en_cut_inf, en_cut_sup)

if save:
    plt.savefig(f'figs/2DHist_E_Dr_Dt_{bins}x{bins}.png', format='png', bbox_inches = 'tight')

plt.show()
```

```python
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sn

import glob
import re
import itertools
import pickle
```

# Load dictionaries with antinu finder results

```python
#Define cuts used for the present results!
posr_cut = 5500

#Primary cuts:
dcFlag_cut = 0x2100000042C2
nhits_cut = 20

# Values selected by looking at the delta_t and delta_r distribution of MC reactor nu data and the fit of the curves
alpha = 9
#Expected values
tau = 215

dr_sup_lim = 1000 #choose a large interval otherwise we are rejecting a lot of events!
dr_inf_lim = 0

#Energy cut on delayed events
energy_delay_inf_cut = 1.8
energy_delay_sup_cut = 3.5
```

```python
bronze_files = 'C:/Users/joanc/jupyter notebooks/Antineutrino Analysis/Real Data Analysis/bronze analysis/antinu_dict/*.pkl'
gold_files = 'C:/Users/joanc/jupyter notebooks/Antineutrino Analysis/Real Data Analysis/gold analysis/antinu_dict/*.pkl'

flist_bronze = glob.glob(bronze_files)
flist_gold = glob.glob(gold_files)

flist = list(itertools.chain(flist_bronze, flist_gold))
flist
```

```python
#data to extract:
var_names = ['energy_prompt', 'energy_delay', 'delta_t', 'delta_r']

# Load data
for i_dx, file_i in enumerate(flist):
    with open(file_i, 'rb') as f:
        locals()['antinu_dict_' + str(i_dx)] = pickle.load(f)

#Concatenate the observables
for var_name_i in var_names:
    locals()[var_name_i] = np.concatenate((antinu_dict_0[var_name_i], 
                                           antinu_dict_1[var_name_i],
                                           antinu_dict_2[var_name_i], 
                                           antinu_dict_3[var_name_i]))
```

# Plots

### Main Title settings

```python
main_title = 'Bronze + Gold Data - '

dcFlag_cut_title = f'mask = {hex(dcFlag_cut)}, '
nhits_cut_title = rf'nHits $\geq$ {nhits_cut}, '
posr_cut_title = rf'R $\leq$ {posr_cut} (mm)'
primary_cuts_title = dcFlag_cut_title + nhits_cut_title + posr_cut_title
secondary_titles = rf'Window $\Delta$t:{alpha}$\tau$ ($\mu$s), $\Delta$r: [{dr_inf_lim}, {dr_sup_lim}] (mm), Energy_delay: [{energy_delay_inf_cut}, {energy_delay_sup_cut}] (MeV)'
```

### Save figs settings

```python
save_dir = 'figs/'
main_title_save = f'Bronze_plus_Gold_alpha_{alpha}_tau_{tau}_en_delay_inf_cut_{energy_delay_inf_cut}_en_delay_sup_cut_{energy_delay_sup_cut}_R_{posr_cut}mm'
```

```python
binwidth_prompt = 0.35
binwidth_delay = 0.12

plt.figure(figsize = (7.5,5))
sn.histplot(energy_prompt, binwidth = binwidth_prompt, color = 'b', element = 'step', alpha = 0, label = f'IBD prompt candidates per {binwidth_prompt} MeV')
sn.histplot(energy_delay, binwidth = binwidth_delay, color = 'r', element = 'step', alpha = 0, label = f'IBD delay candidates per {binwidth_delay} MeV')
plt.legend(loc = 'best', fontsize = 9)

plt.xlabel('Energy (MeV)')

#plt.ylabel(f'Counts/{binwidth} (MeV)')
#plt.yscale('log')

plt.title(main_title + primary_cuts_title + '\n' + secondary_titles, fontsize = 10)
plt.savefig(save_dir + 'energy_' + main_title_save+'.png', format = 'png', bbox_inches = 'tight')

plt.show()
```

```python
plt.figure(figsize = (7.5,5))

sn.histplot(delta_t, bins = 14, color = 'b', element = 'step', alpha = 0)

plt.xlabel(r'$\Delta$t ($\mu$s)')
plt.ylabel('Candidate Count')
plt.title(main_title + primary_cuts_title + '\n' + secondary_titles, fontsize = 10)

#plt.savefig(save_dir + 'delta_t_' + main_title_save+'.png', format = 'png', bbox_inches = 'tight')
plt.show()
```

```python
plt.figure(figsize = (7.5,5))

sn.histplot(delta_r, bins = 10, color = 'b', element = 'step', alpha = 0)

plt.xlabel(r'$\Delta$r (mm)')
plt.ylabel('Candidate Count')
plt.title(main_title + primary_cuts_title + '\n' + secondary_titles, fontsize = 10)

#plt.savefig(save_dir + 'delta_r_' + main_title_save +'.png', format = 'png', bbox_inches = 'tight')
plt.show()
```

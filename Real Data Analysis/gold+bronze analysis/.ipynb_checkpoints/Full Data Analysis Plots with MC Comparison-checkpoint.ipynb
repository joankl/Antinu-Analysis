```python
import colorsys
# Función para generar variantes más claras y más oscuras de un color
def generate_color_variants(base_color, num_variants, factor = 0.3):
    # Convertir el color base a valores de HSV
    r, g, b = base_color
    h, l, s = colorsys.rgb_to_hls(r, g, b)
    
    # Generar variantes
    variants = []
    for i in range(1, num_variants + 1):
        # Hacer el color más claro
        lighter = colorsys.hls_to_rgb(h, min(1, l * (1 + factor * i)), s)
        # Hacer el color más oscuro
        darker = colorsys.hls_to_rgb(h, max(0, l * (1 - factor * i)), s)
        variants.append(lighter)
        variants.append(darker)
        
    return variants

# Color base (rojo)
base_red = (1, 0, 0) # RGB para rojo puro
# Color base (azuk)
base_blue = (0, 0, 1) # RGB para azul puro
num_variants = 5

red_variants = generate_color_variants(base_red, num_variants)
blue_variants = generate_color_variants(base_blue, num_variants)
```

# Useful Functions

```python
def posr_cal(x, y, z):
    dr = np.sqrt(x**2 + y**2 + z**2)
    return dr
```

```python
#Primary cuts:
dcFlag_cut = 0x2100000042C2
nhits_cut = 20

#Define the cut values used to construct the dictionaries to select the data.pkl to read --------------------------------------
en_cut_inf = 1.2
en_cut_sup = 8.0
posr_cut_sup = 6000
runID_cut = 300700

#Vetoing HS and atmospheric :
nhits_hs_number = 1500  #nhits above this value is an hotspot suspected
nhits_atm_number = 3000 #nhits above this value is an atmospheric suspected
dt_hs_remove = 1e6  # Remove 1s of data after a hs
dt_atm_remove = 20*1e6  # Remove 20s of data after an atmospheric


# Antinufinder cuts
dt_sup_lim = 1000
dt_inf_lim = 0.5

dr_sup_lim = 1000 
dr_inf_lim = 0

energy_delay_inf_cut = 1.9
energy_delay_sup_cut = 2.3

# Main Data Directory -------------------------------------------------
main_dir = 'C:/Users/joanc/jupyter notebooks/Antineutrino Analysis/Real Data Analysis/Accidentals Analysis/data out dictionaries/'

dict_bronze_fname = f'Accidentals_bronze_dict_delta_t_{dt_inf_lim}_{dt_sup_lim}(mcs)_dr_{dr_inf_lim}_{dr_sup_lim}(mm)_en_{en_cut_inf}_{en_cut_sup}(MeV)_en_delay_{energy_delay_inf_cut}_{energy_delay_sup_cut}(MeV)'
dict_gold_fname = f'Accidentals_gold_dict_delta_t_{dt_inf_lim}_{dt_sup_lim}(mcs)_dr_{dr_inf_lim}_{dr_sup_lim}(mm)_en_{en_cut_inf}_{en_cut_sup}(MeV)_en_delay_{energy_delay_inf_cut}_{energy_delay_sup_cut}(MeV)'

# Load Accidentals Dictionaries ---------------------------------------
with open(main_dir + dict_bronze_fname + '.pkl', 'rb') as f:
    dict_acc_bronze = pickle.load(f)

with open(main_dir + dict_gold_fname + '.pkl', 'rb') as f:
    dict_acc_gold = pickle.load(f)

# Observables selection -----------------------------------------------

posx_prompt = np.append(dict_acc_bronze['posx_prompt'], dict_acc_gold['posx_prompt'])
posy_prompt = np.append(dict_acc_bronze['posy_prompt'], dict_acc_gold['posy_prompt'])
posz_prompt = np.append(dict_acc_bronze['posz_prompt'], dict_acc_gold['posz_prompt'])
posr_prompt = posr_cal(posx_prompt, posy_prompt, posz_prompt)

posx_delay = np.append(dict_acc_bronze['posx_delay'], dict_acc_gold['posx_delay'])
posy_delay = np.append(dict_acc_bronze['posy_delay'], dict_acc_gold['posy_delay'])
posz_delay = np.append(dict_acc_bronze['posz_delay'], dict_acc_gold['posz_delay'])
posr_delay = posr_cal(posx_delay, posy_delay, posz_delay)

energy_prompt_acc = np.append(dict_acc_bronze['energy_prompt'], dict_acc_gold['energy_prompt'])
energy_delay_acc = np.append(dict_acc_bronze['energy_delay'], dict_acc_gold['energy_delay'])
dt_acc = np.append(dict_acc_bronze['delta_t'], dict_acc_gold['delta_t'])
dr_acc = np.append(dict_acc_bronze['delta_r'], dict_acc_gold['delta_r'])

# Data cuts ------------------------------------------------------------

# dt cut
dt_sup_lim = 900
dt_condition = (dt_acc <= dt_sup_lim)

posr_prompt = posr_prompt[dt_condition]
posr_delay = posr_delay[dt_condition]
energy_prompt_acc = energy_prompt_acc[dt_condition]
energy_delay_acc = energy_delay_acc[dt_condition]
dt_acc = dt_acc[dt_condition]
dr_acc = dr_acc[dt_condition]

N_initial_acc = len(energy_prompt_acc) #Number of intial accidentals before posr cut

# RadialPosition cut
posr_condition = (posr_prompt <= posr_cut_sup) & (posr_delay <= posr_cut_sup)

energy_prompt_acc = energy_prompt_acc[posr_condition]
energy_delay_acc = energy_delay_acc[posr_condition]
dt_acc = dt_acc[posr_condition]
dr_acc = dr_acc[posr_condition]

N_final_acc = len(energy_prompt_acc) #Number of final accidentals after posr cut
```

# Scaling of MC Data to Real Data

Obtain the Scaling Factors to Normalize the MC counts of Reactor Nu, Geonu and alpha_n. These factor are obtained knowing:
- The rate of expected events per seleciont efficiency of cut per (day)^-{1}
- The livetime of the real dataset
- The number of simulated events

```python
# ========= Definition of Important Quantities =========

# General quantities:
dataset_tot_lf = 228.04 # Nº of Days of the net time real dataset (without applying cuts to find coincidence events)
cut_time = 27.82 # Nº of days to be removed from the dataset_tot_lf due to run cuts, hotsposts and atmospherics
total_lf = dataset_tot_lf - cut_time  # Total livetime of the raw dataset

# Event Rate = N_evs_expected*efficiency/day:
N_reactor_per_day = 35.8/134.4
N_alpha_n_per_day = 27.3/134.4
N_geonu_per_day = 9.5/134.4
N_acc_per_day = 51.6/134.4
#N_tot_per_day = N_reactor_per_day + N_alpha_n_per_day + N_geonu_per_day

# Nº of MC pairs events:
N_mc_reactor = len(energy_prompt_reactor_nu)
N_mc_reactor_osc = len(energy_prompt_reactor_nu_osc)
N_mc_geonu = len(energy_prompt_geo_nu)
N_mc_alpha_n = len(energy_prompt_alpha_n)
N_mc_bipo214 = len(energy_prompt_bipo214)
#N_mc_tot = N_mc_reactor_osc + N_mc_geonu + N_mc_alpha_n

# Efficiency of MC cuts:
eff_mc_cut_reactor = N_mc_reactor/np.sum(initial_pairs_ev_reactor_nu)
eff_mc_cut_reactor_osc = N_mc_reactor_osc/np.sum(initial_pairs_ev_reactor_nu_osc)
eff_mc_cut_geo = N_mc_geonu/np.sum(initial_pairs_ev_geo_nu)
eff_mc_cut_alpha_n = N_mc_alpha_n/np.sum(initial_pairs_ev_alpha_n)
eff_mc_cut_bipo214 = N_mc_bipo214/np.sum(initial_pairs_ev_bipo214)
eff_cut_acc = N_final_acc/N_initial_acc
#eff_mc_tot = eff_mc_cut_reactor_osc + eff_mc_cut_geo + eff_mc_cut_alpha_n
#eff_mc_tot = ( N_mc_reactor_osc + N_mc_geonu + eff_mc_cut_alpha_n) / (np.sum(initial_pairs_ev_reactor_nu_osc) + np.sum(initial_pairs_ev_geo_nu) + np.sum(initial_pairs_ev_alpha_n))

print(f'Reactor Nu MC cuts Selection Efficiency = {eff_mc_cut_reactor}')
print(f'Reactor Nu Osc. MC cuts Selection Efficiency = {eff_mc_cut_reactor_osc}')
print(f'Geo Nu MC cuts Selection Efficiency = {eff_mc_cut_geo}')
print(f'alpha_n MC cuts Selection Efficiency = {eff_mc_cut_alpha_n}')
print(f'BiPo214 MC cuts Selection Efficiency = {eff_mc_cut_bipo214}')
print(f'Accidentals cuts Selection Efficiency = {eff_cut_acc}')

# Scaled events and scale factors Computation:
N_scaled_mc_reactor = total_lf * N_reactor_per_day * eff_mc_cut_reactor
N_scaled_mc_reactor_osc = total_lf * N_reactor_per_day * eff_mc_cut_reactor_osc
N_scaled_mc_geonu = total_lf * N_geonu_per_day * eff_mc_cut_geo
N_scaled_mc_alpha_n = total_lf * N_alpha_n_per_day * eff_mc_cut_alpha_n
N_scaled_acc = total_lf * N_acc_per_day * eff_cut_acc
#N_scaled_mc_tot = total_lf * N_tot_per_day

scale_factor_reactor = (N_scaled_mc_reactor / N_mc_reactor)
scale_factor_reactor_osc = (N_scaled_mc_reactor_osc / N_mc_reactor_osc) 
scale_factor_geonu = (N_scaled_mc_geonu / N_mc_geonu) * eff_mc_cut_geo
scale_factor_alpha_n = (N_scaled_mc_alpha_n / N_mc_alpha_n) 
scale_factor_acc = (N_scaled_acc / N_final_acc) 
#scale_factor_total_mc = (N_scaled_mc_tot / N_mc_tot) 

print('====================================================================')
print(f'Reactor Nu = {N_mc_reactor} MC events scaled to {N_scaled_mc_reactor}')
print(f'Reactor Nu Osc. = {N_mc_reactor_osc} MC events scaled to {N_scaled_mc_reactor_osc}')
print(f'Geo Nu = {N_mc_geonu} MC events scaled to {N_scaled_mc_geonu}')
print(f'alpha-n = {N_mc_alpha_n} MC events scaled to {N_scaled_mc_alpha_n}')
print(f'Accidenals = {N_final_acc} events scaled to {N_scaled_acc}')
```

```python
#Function to Fit
def fit_exp(x, n0, lambda_):
    return n0 * np.exp(lambda_ * x)
```

```python
# ================== Plot Settings ==================

font_style_title = {'family':'serif', 'weight': 'normal','color':'black','size':13}
font_style_axis= {'family':'serif', 'weight': 'normal','color':'black','size':12}

font_prop = font_manager.FontProperties(family=font_style_axis['family'], weight=font_style_axis['weight'], size=font_style_axis['size'])

linewidth = 1.2
alpha = 0.2

# Binwidths
binwidth_dt = 50      # μs
binwidth_dr = 80     # mm

# Definir bin edges comunes y convertir a lista para evitar error con seaborn
dt_bins = list(np.arange(
    min(delta_t_real_data),
    max(delta_t_real_data) + binwidth_dt,
    binwidth_dt))

dr_bins = list(np.arange(
    min(delta_r_real_data),
    max(delta_r_real_data) + binwidth_dr,
    binwidth_dr))

fig, axes = plt.subplots(1, 2, figsize=(24, 7.5))

# ======= TOTAL MC COMPUTATION =========

# ------------------------- Dt histogram components of scaled MC -------------------------
counts_dt_reactor_osc, _ = np.histogram(dt_reactor_nu_osc, bins = dt_bins, weights = np.ones(len(dt_reactor_nu_osc)) * scale_factor_reactor_osc)
counts_dt_geo, _ = np.histogram(dt_geo_nu, bins = dt_bins, weights = np.ones(len(dt_geo_nu)) * scale_factor_geonu)
counts_dt_alpha_n, _ = np.histogram(dt_alpha_n, bins = dt_bins,weights = np.ones(len(dt_alpha_n)) * scale_factor_alpha_n)
counts_dt_acc, _ = np.histogram(dt_acc, bins = dt_bins, weights = np.ones(len(dt_acc)) * scale_factor_acc)
# Histogram sum
counts_dt_mc_total = counts_dt_reactor_osc + counts_dt_geo + counts_dt_alpha_n + counts_dt_acc

# ------------------------- Dr histogram components of scaled MC -------------------------
counts_dr_reactor_osc, _ = np.histogram(dr_reactor_nu_osc, bins = dr_bins, weights = np.ones(len(dr_reactor_nu_osc)) * scale_factor_reactor_osc)
counts_dr_geo, _ = np.histogram(dr_geo_nu, bins = dr_bins, weights = np.ones(len(dr_geo_nu)) * scale_factor_geonu)
counts_dr_alpha_n, _ = np.histogram(dr_alpha_n, bins = dr_bins,weights = np.ones(len(dr_alpha_n)) * scale_factor_alpha_n)
counts_dr_acc, _ = np.histogram(dr_acc, bins = dr_bins, weights = np.ones(len(dr_acc)) * scale_factor_acc)
# Histogram sum
counts_dr_mc_total = counts_dr_reactor_osc + counts_dr_geo + counts_dr_alpha_n + counts_dr_acc

# =========================
# Δt (µs)
# =========================
counts_dt, edges_dt = np.histogram(delta_t_real_data, bins=dt_bins)
bin_centers_dt = (edges_dt[:-1] + edges_dt[1:]) / 2

errors_dt = np.sqrt(counts_dt)

#Remove zero count points--------------------
bin_centers_dt_real_data = bin_centers_dt[counts_dt>0]
errors_dt = errors_dt[counts_dt>0]
counts_dt = counts_dt[counts_dt>0]
# -------------------------------------------

axes[0].errorbar(bin_centers_dt_real_data, counts_dt, yerr = errors_dt, fmt='o', 
                 color='black', label = 'Real data (Stat. Unc.)', elinewidth=1, capsize=2)
axes[0].plot(x_t, y_fit, linewidth = linewidth, color = '#fc0303', 
             label = r'$N_{0}e^{-\tau\Delta\text{t}}$ Fit' + rf' $\tau = ({tau:.0f} \pm {tau_err:.0f}$) $\mu$s')

sn.histplot(x=dt_reactor_nu_osc, bins = dt_bins, weights = np.ones(len(dt_reactor_nu_osc)) * scale_factor_reactor_osc,
            element='step', alpha = alpha, linewidth = linewidth,
            color='#f0058e', ax = axes[0], label=r'Reactor $\bar{\nu}$ MC')

sn.histplot(x=dt_geo_nu, bins=dt_bins, weights=np.ones(len(dt_geo_nu)) * scale_factor_geonu,
            element='step', alpha=alpha, linewidth=linewidth,
            color='#05f03c', ax=axes[0], label=r'Geo-$\bar{\nu}$ MC')

sn.histplot(x=dt_alpha_n, bins=dt_bins,  weights=np.ones(len(dt_alpha_n)) * scale_factor_alpha_n, element='step', alpha=alpha, linewidth=linewidth,
    color='#051cf0', ax=axes[0], label=r'C-13 $\alpha$-n MC')

#sn.histplot(x=dt_acc, bins=dt_bins,  weights=np.ones(len(dt_acc)) * scale_factor_acc, element='step', alpha = 0, linewidth=linewidth,
#    color='black', linestyle='--', ax=axes[0], label=r'Accidentals')

axes[0].hist(bin_centers_dt, bins = dt_bins, weights = counts_dt_mc_total, histtype='step',color='#4f4a52', linewidth = 1.5, label='Total MC')
axes[0].set_xlabel(r'$\Delta$t ($\mu$s)', fontdict = font_style_axis)
axes[0].set_ylabel(f'Counts per {binwidth_dt} µs', fontdict = font_style_axis)
axes[0].set_xlim(edges_dt[0], edges_dt[-1])
#axes[0].set_yscale('log')

# =========================
# Δr (mm)
# =========================
counts_dr, edges_dr = np.histogram(delta_r_real_data, bins=dr_bins)
bin_centers_dr = (edges_dr[:-1] + edges_dr[1:]) / 2

errors_dr = np.sqrt(counts_dr)
axes[1].errorbar(bin_centers_dr, counts_dr, yerr = errors_dr, fmt='o', 
                 color='black', label = 'Real data (Stat. Unc.)', elinewidth=1, capsize=2)

#sn.histplot(x=dr_reactor_nu, bins=dr_bins, weights=np.ones(len(dr_reactor_nu)) * scale_factor_reactor,
#            element='step', alpha=alpha, linewidth=linewidth, 
#            color='#f00511', ax=axes[1], label=r'Reactor $\bar{\nu}$ MC')

sn.histplot(x=dr_reactor_nu_osc, bins=dr_bins, weights=np.ones(len(dr_reactor_nu_osc)) * scale_factor_reactor_osc,
            element='step', alpha=alpha, linewidth=linewidth,
            color='#f0058e', ax=axes[1], label=r'Reactor $\bar{\nu}$ MC')

sn.histplot(x=dr_geo_nu, bins=dr_bins, weights=np.ones(len(dr_geo_nu)) * scale_factor_geonu,
            element='step', alpha = alpha, linewidth=linewidth, 
            color='#05f03c', ax=axes[1], label=r'Geo-$\bar{\nu}$ MC')

sn.histplot(x=dr_alpha_n, bins=dr_bins, weights=np.ones(len(dr_alpha_n)) * scale_factor_alpha_n,
            element='step', alpha=alpha, linewidth=linewidth,
            color='#051cf0', ax=axes[1], label=r'C-13 $\alpha$-n MC')

#sn.histplot(x = dr_acc, bins = dr_bins, weights = np.ones(len(dr_acc)) * scale_factor_acc,
#            element = 'step', alpha = 0, linewidth = linewidth,
#            color = 'black', linestyle = '--', ax = axes[1], label = r'Accidentals')

axes[1].hist(bin_centers_dr, bins = dr_bins, weights = counts_dr_mc_total, histtype='step',color='#4f4a52', linewidth = 1.5, label='Total MC')
axes[1].set_xlabel(r'$\Delta$r (mm)', fontdict = font_style_axis)
axes[1].set_ylabel(f'Counts per {binwidth_dr} mm', fontdict = font_style_axis)
axes[1].set_xlim(edges_dr[0], edges_dr[-1])
#axes[1].set_yscale('log')

#Common legends :
# Obtain the handles and labels of only one plot (just because they share labels!)
handles, labels = axes[0].get_legend_handles_labels()

# Create one legent aout of the axis plots
fig.legend(handles, labels,
           loc = 'upper right',        
           bbox_to_anchor = (0.94, 0.893),  # coordinates of the box at the right (x, y)
           frameon = True,             # white background
           edgecolor='black',
           prop = font_prop,
           labelcolor = font_style_axis['color'])

# Adjunts the plot size to left space to the legend box 
fig.subplots_adjust(right = 0.8)  # right space

plt.suptitle( main_title + general_cuts_title + '\n' + vetoing_title + '\n' + antinufinder_title, fontdict = font_style_title)

if save:
    plt.savefig(save_dir + 'dt_dr_' + main_title_save + '.png', dpi=300, bbox_inches='tight')
    
plt.show()
```
